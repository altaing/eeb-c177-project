# Commented Function and Subfunctions

```python
def check(file): # defining function "check"
    import re # imports regex
    
    csv = re.compile(r'.*\.csv') # will search for input ending in .csv
    file = csv.search(file) # returns first match
    return bool(file) # if the input ends in .csv, this function will return True

def scatterbigy(df): # defining function "scatterbigy"
    import seaborn as sns # importing seaborn
    from scipy.stats import pearsonr # importing pearsonr
    
    avg = df.mean(axis=0) # average function
    mx = df.max(axis=0) # max function
    mn = df.min(axis=0) # min function
    stdev = df.std(axis=0) # standard deviation function
    v1, v2 = df # splitting df into two varibles
    sns.lmplot(v1, v2, data=df,fit_reg=True) # plotting a scatterplot using seaborn  
    sns.jointplot(v1, v2, data=df, alpha = 0.1, stat_func=pearsonr, color='green') # creates another plot with statistical analyses
    
    print("Average Values:\n", avg, "\n") # prints out the average
    print("Max Values:\n", mx, "\n") # prints out the max
    print("Minimum Values:\n", mn, "\n") # prints out the minimum
    print("Standard Deviation:\n", stdev, "\n") # prints out the standard deviation

def boxbigy(df): # defining function "boxbigy"
    import seaborn as sns # importing seaborn
    from scipy.stats import pearsonr # importing pearsonr
    import matplotlib.pyplot as plt #importing matplotlib.pyplot
    
    avg = df.mean(axis=0) # average function
    mx = df.max(axis=0) # max function
    mn = df.min(axis=0) # min function
    stdev = df.std(axis=0) # standard deviation function
    v1, v2 = df # splitting df into two variables
    
    plt.figure(figsize=(14,6)) # adjusting figure size
    sns.boxplot(v1,v2, data=df) # plotting a boxplot using seaborn
    sns.jointplot(v1, v2, data=df, alpha = 0.1, stat_func=pearsonr, color='green') # creates plot with statistical analyses
    
    print("Average Values:\n", avg, "\n") # prints out the average
    print("Max Values:\n", mx, "\n") # prints out the max
    print("Minimum Values:\n", mn, "\n") # prints out the minimum
    print("Standard Deviation:\n", stdev, "\n") # prints out the standard deviation
    
def scatterbigx(df): # defining function "scatterbigx"
    import seaborn as sns # importing seaborn
    from scipy.stats import pearsonr # importing pearsonr
    
    avg = df.mean(axis=0) # average function
    mx = df.max(axis=0) # max function
    mn = df.min(axis=0) # min function
    stdev = df.std(axis=0) # standard deviation function
    v1, v2 = df # splitting df into two variables
    
    sns.lmplot(v2, v1, data=df,fit_reg=True) # plotting scatterplot using seaborn
    sns.jointplot(v2, v1, data=df, alpha = 0.1, stat_func=pearsonr, color='green') # creates plot with statistical analyses
    
    print("Average Values:\n", avg, "\n") # prints out the average
    print("Max Values:\n", mx, "\n") # prints out the max
    print("Minimum Values:\n", mn, "\n") # prints out the minimum
    print("Standard Deviation:\n", stdev, "\n") # prints out the standard deviation
    
def boxbigx(df): # defining function "boxbigx"
    import seaborn as sns # importing seaborn
    from scipy.stats import pearsonr # importing pearsonr
    import matplotlib.pyplot as plt # importing matplotlib.pyplot
    
    avg = df.mean(axis=0) # average function
    mx = df.max(axis=0) # max function
    mn = df.min(axis=0) # min function
    stdev = df.std(axis=0) # standard deviation function
    v1, v2 = df # splitting df into two variables
    
    plt.figure(figsize=(14,6)) # adjusting figure size
    sns.boxplot(v2,v1, data=df) # plotting boxplot using seaborn
    sns.jointplot(v2, v1, data=df, alpha = 0.1, stat_func=pearsonr, color='green') # creates plot with statistical analyses
    
    print("Average Values:\n", avg, "\n") # prints out the average
    print("Max Values:\n", mx, "\n") # prints out the max
    print("Minimum Values:\n", mn, "\n") # prints out the minimum
    print("Standard Deviation:\n", stdev, "\n") # prints out the standard deviation

def boxeq(df): # defining function "boxeq"
    import seaborn as sns # importing seaborn
    
    avg = df.mean(axis=0) # average function
    mx = df.max(axis=0) # max function
    mn = df.min(axis=0) # min function
    stdev = df.std(axis=0) # standard deviation function
    
    sns.boxplot(df, data=df) # plotting boxplot using seaborn
    
    print("Average Values:\n", avg, "\n") # prints out the average
    print("Max Values:\n", mx, "\n") # prints out the max
    print("Minimum Values:\n", mn, "\n") # prints out the minimum
    print("Standard Deviation:\n", stdev, "\n") # prints out the standard deviation

def histeq(df): # defining function "histeq"
    import seaborn as sns # importing seaborn
    
    avg = df.mean(axis=0) # average function
    mx = df.max(axis=0) # max function
    mn = df.min(axis=0) # min function
    stdev = df.std(axis=0) # standard deviation function
    modee = df.mode(axis=0) # mode function
    
    sns.distplot(df) # plotting histogram using seaborn
    
    print("Average Values:\n", avg, "\n") # prints out the average
    print("Max Values:\n", mx, "\n") # prints out the max
    print("Minimum Values:\n", mn, "\n") # prints out the minimum
    print("Standard Deviation:\n", stdev, "\n") # prints out the standard deviation
    print("Mode:\n", modee) # prints out the mode
    
def bigy(filename, clmns): # defining function "bigy"
    import pandas as pd # importing pandas
    df = pd.read_csv(filename, usecols = clmns) # importing file and designating columns to use from pandas
    vis = str(input('What type of plot would you like to see?\nType "scatter" or "boxplot":\n')) # allows user input for desired plot
    print('\n') # blank space for visual clarity
    if vis == "scatter": # setting up if statement
        scatterbigy(df) # previously defined function
    if vis == "boxplot": # other if statement arugment
        boxbigy(df) # previously defined function
        
def bigx(filename, clmns): # defining function "bigx"
    import pandas as pd # importing pandas
    df = pd.read_csv(filename, usecols = clmns) # importing file and designating columns to use from pandas
    vis = str(input('What type of plot would you like to see?\nType "scatter" or "boxplot":\n')) # allows user input for desired plot
    print('\n') # blank space for visual clarity
    if vis == "scatter": # setting up if statement
        scatterbigx(df) # previously defined function 
    if vis == "boxplot": # other if statement argument
        boxbigx(df) # previously defined function
    
def eqxy(filename, clmns): # defining function "eqxy"
    import pandas as pd #importing pandas
    df = pd.read_csv(filename, usecols = clmns) # importing file and designating columns to use from pandas
    vis = str(input('What plot of plot would you like to see?\nType "boxplot", or "histogram":\n')) # allows user input for desired plot
    print('\n') # blank space for visual clarity
    if vis == "boxplot": # setting up if statement
        boxeq(df) # previously defined function
    if vis == "histogram": # other if statement argument
        histeq(df) # previously defined function
        
filename = '' # defining "filename"
clmns = [] # defining "clmns"
vis = '' # defining "vis"

def finalfunction(filename, clmns, vis): # defining function "finalfunction"
    print('Input .csv file you wish to analyze:') # prints instructions for user
    filename = str(input()) # allows for a string input for "filename"
    assert check(filename) == True, 'This function will only work with .csv files\nPlease input a .csv file.' # assert statement that ensures the user input is of the correct file type
    print('\n') # blank space for visual clarity
    cl = [] # defining "cl"
    clmns = cl # setting "clmns" equa to "cl"
    print('Key:\nage = 0, sex = 1\nchest pain = 2, resting blood pressure = 3\ncholesterol = 4, fasting blood sugar = 5\nresting ECG = 6, max heart rate = 7\nexercise induced angina = 8, ST depression induced by excercise = 9\nslope of the peak exercise ST segment = 10\nnumber of major vessels colored by flourosopy = 11\nthal = 12, target = 13') # providing user with a key for all elements in the heart.csv file
    print('\n') # blank space for visual clarity
    for i in range(0, 2): # setting up for loop
        print("Enter the number that corresponds to the desired element:") # prints instructions for user
        item = int(input()) # allows for integer inputs
        assert item <= 13, 'Please only input values that are assciated with defined elements.' # asserts that the user input does not exceed the elements within the heart.csv file
        cl.append(item) # inputs user input in the previously defined empty list
    x, y = cl # splits "cl" into two variables
    print('\n') # blank space for visual clarity
    if x < y: # setting up if statement
        bigy(filename, clmns) # previously defined function
    elif x > y: # one if statement argument
        bigx(filename, clmns) # previously defined function
    elif x == y: # other if statement argument
        eqxy(filename, clmns) #previously defined function
```
